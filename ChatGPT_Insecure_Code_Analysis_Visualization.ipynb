## Quality Metrics

```python
import pandas as pd

# Mean_Metrics_Curr
df = pd.read_csv('LLMSecEval-prompts-analyze.csv')

mean_naturalness_curr = df['our_Naturalness'].mean()
mean_expressiveness_curr = df['our_Expressiveness'].mean()
mean_adequacy_curr = df['our_Adequacy'].mean()
mean_conciseness_curr = df['our_Conciseness'].mean()

# Mean_Metrics_Old
mean_naturalness_old = df['their_Naturalness'].mean()
mean_expressiveness_old = df['their_Expressiveness'].mean()
mean_adequacy_old = df['their_Adequacy'].mean()
mean_conciseness_old = df['their_Conciseness'].mean()
```

```python
# Naturalness
curr_naturalness_c = df.loc[df['Language'] == 'C', 'our_Naturalness'].mean()
curr_naturalness_python = df.loc[df['Language'] == 'Python', 'our_Naturalness'].mean()
old_naturalness_c = df.loc[df['Language'] == 'C', 'their_Naturalness'].mean()
old_naturalness_python = df.loc[df['Language'] == 'Python', 'their_Naturalness'].mean()

# Expressiveness
curr_expressiveness_c = df.loc[df['Language'] == 'C', 'our_Expressiveness'].mean()
curr_expressiveness_python = df.loc[df['Language'] == 'Python', 'our_Expressiveness'].mean()
old_expressiveness_c = df.loc[df['Language'] == 'C', 'their_Expressiveness'].mean()
old_expressiveness_python = df.loc[df['Language'] == 'Python', 'their_Expressiveness'].mean()

# Adequacy
curr_adequacy_c = df.loc[df['Language'] == 'C', 'our_Adequacy'].mean()
curr_adequacy_python = df.loc[df['Language'] == 'Python', 'our_Adequacy'].mean()
old_adequacy_c = df.loc[df['Language'] == 'C', 'their_Adequacy'].mean()
old_adequacy_python = df.loc[df['Language'] == 'Python', 'their_Adequacy'].mean()

# Conciseness
curr_conciseness_c = df.loc[df['Language'] == 'C', 'our_Conciseness'].mean()
curr_conciseness_python = df.loc[df['Language'] == 'Python', 'our_Conciseness'].mean()
old_conciseness_c = df.loc[df['Language'] == 'C', 'their_Conciseness'].mean()
old_conciseness_python = df.loc[df['Language'] == 'Python', 'their_Conciseness'].mean()
```

```python
import matplotlib.pyplot as plt

data = {'Past': [mean_naturalness_old, mean_expressiveness_old, mean_adequacy_old, mean_conciseness_old],
        'Current': [mean_naturalness_curr,mean_expressiveness_curr, mean_adequacy_curr, mean_conciseness_curr]}
df = pd.DataFrame(data, columns = ['Past', 'Current'], index = ['Naturalness', 'Expressiveness', 'Adequacy', 'Conciseness'])

ax = df.plot.bar(color = ['#B4B4B4', '#4DED30'], figsize = (10, 7))

# Annotation
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')
    
plt.xlabel("Metrics", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.title("Mean Score of Metrics: Past vs. Current", fontsize=15)
plt.show()
```

## Quality Metrics : C vs Python

```python
data = {'Past': [mean_naturalness_old, mean_expressiveness_old, mean_adequacy_old, mean_conciseness_old],
        'Current': [mean_naturalness_curr,mean_expressiveness_curr, mean_adequacy_curr, mean_conciseness_curr]}
df = pd.read_csv('LLMSecEval-prompts-analyze.csv')

# Naturalness
curr_naturalness_c = df.loc[df['Language'] == 'C', 'our_Naturalness'].mean()
curr_naturalness_python = df.loc[df['Language'] == 'Python', 'our_Naturalness'].mean()
old_naturalness_c = df.loc[df['Language'] == 'C', 'their_Naturalness'].mean()
old_naturalness_python = df.loc[df['Language'] == 'Python', 'their_Naturalness'].mean()

# Expressiveness
curr_expressiveness_c = df.loc[df['Language'] == 'C', 'our_Expressiveness'].mean()
curr_expressiveness_python = df.loc[df['Language'] == 'Python', 'our_Expressiveness'].mean()
old_expressiveness_c = df.loc[df['Language'] == 'C', 'their_Expressiveness'].mean()
old_expressiveness_python = df.loc[df['Language'] == 'Python', 'their_Expressiveness'].mean()

# Adequacy
curr_adequacy_c = df.loc[df['Language'] == 'C', 'our_Adequacy'].mean()
curr_adequacy_python = df.loc[df['Language'] == 'Python', 'our_Adequacy'].mean()
old_adequacy_c = df.loc[df['Language'] == 'C', 'their_Adequacy'].mean()
old_adequacy_python = df.loc[df['Language'] == 'Python', 'their_Adequacy'].mean()

# Conciseness
curr_conciseness_c = df.loc[df['Language'] == 'C', 'our_Conciseness'].mean()
curr_conciseness_python = df.loc[df['Language'] == 'Python', 'our_Conciseness'].mean()
old_conciseness_c = df.loc[df['Language'] == 'C', 'their_Conciseness'].mean()
old_conciseness_python = df.loc[df['Language'] == 'Python', 'their_Conciseness'].mean()
```

```python
categories = ['Naturalness', 'Expressiveness', 'Adequacy', 'Conciseness']
subcategories = ['C', 'Python']
time = ['Past', 'Current']

# Define data for the bars
data = {
    'C': {
        'Past': [old_naturalness_c, old_expressiveness_c, old_adequacy_c, old_conciseness_c],
        'Current': [curr_naturalness_c, curr_expressiveness_c, curr_adequacy_c, curr_conciseness_c]
    },
    'Python': {
        'Past': [old_naturalness_python, old_expressiveness_python, old_adequacy_python, old_conciseness_python],
        'Current': [curr_naturalness_python, curr_expressiveness_python, curr_adequacy_python, curr_conciseness_python]
    }
}

# Plotting
bar_width = 0.2
index = range(len(categories))
colors = ['#FC8EAC', '#FF007F', '#A3D5FF','navy']
counter = 0
fig, ax = plt.subplots(figsize = (12, 9))

for i, subcategory in enumerate(subcategories):
    for j, section in enumerate(time):
        bars = [data[subcategory][section][k] for k in range(len(categories))]
        extra_space = 0.2 
        ax.bar([x + (i * bar_width * (len(time))) + (j * bar_width) for x in index], 
               bars, width=bar_width, label=f'{subcategory} - {section}',
              color = colors[counter])
        counter += 1

ax.set_xlabel('Metrics', fontsize=12)
ax.set_ylabel('Score Mean', fontsize=12)
ax.set_title('C and Python Metrics Score Mean: Past vs. Current', fontsize=15)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

extra_space_after_set = len(categories) * bar_width * len(time) * extra_space
ax.set_xticks([x + ((len(subcategories) - 1) / 2) * bar_width + extra_space_after_set for x in index])
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

## Vulnerability

```python
df = pd.read_csv('final_LLMSecEval-prompts-analyze.csv')

# Count True and False
count_vulnerability = df['Vulnerable'].value_counts()
count_our_vulnerability = df['our_Vulnerable'].value_counts()

bar_width = 0.05
index = range(2)

fig, ax = plt.subplots(figsize=(5, 8))

# True (Red)
ax.bar(index, [count_vulnerability[True], count_our_vulnerability[True]], label='True', color='red')

# False (Green)
ax.bar(index, [count_vulnerability[False], count_our_vulnerability[False]],
       bottom=[count_vulnerability[True], count_our_vulnerability[True]], label='False', color='green')

for i, value in enumerate([count_vulnerability[True], count_our_vulnerability[True]]):
    ax.text(i, value / 2, str(value), ha='center', va='center', color='white', fontsize=12)

for i, value in enumerate([count_vulnerability[False], count_our_vulnerability[False]]):
    ax.text(i, count_vulnerability[True] + value / 2, str(value), ha='center', va='center', color='black', fontsize=12)

# Labels
ax.set_xlabel('Data')
ax.set_ylabel('Count')
ax.set_title('Number of Vulnerable Codes')
ax.set_xticks(index)
ax.set_xticklabels(['Past', 'Current'])
ax.legend()

plt.show()
```
